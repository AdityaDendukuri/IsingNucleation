```python
import numpy as np 
import matplotlib.pyplot as plt
mag=np.loadtxt("mag.txt")
```

```python
# reshape data based on number of windows
n_windows = 30
mag1 = mag.reshape([n_windows, -1])
mag1.shape
```

## Histograms of nucleus size in each simulation window

```python
fig= plt.figure(figsize=(10,3))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
x = 15
for i in range(n_windows):
    axes.hist(mag1[i], alpha=0.5, bins=10)
    #plt.axvline(mag1[i].mean(), color='k', linestyle='dashed', linewidth=1)

axes.set_xlabel("N")
plt.title("histogram of nucleus size for each overlapping window")
plt.savefig("window_histograms_raw.pdf")
```

## Conveting those histograms to free energies (-ln(freq ))

```python
fig= plt.figure(figsize=(10,3))

axes= fig.add_axes([0.1,0.1,0.8,0.8])
x = 15
for i in range(1, n_windows):
    #axes.hist(mag1[i], alpha=0.5, bins=10)
    #plt.axvline(mag1[i].mean(), color='k', linestyle='dashed', linewidth=1)
    hist = np.histogram(mag1[i])
    loghist=-np.log(hist[0])
    plt.plot(hist[1][:-1], loghist)
    x+=15
axes.set_xlabel("N")
plt.savefig("window_free_energies_raw.pdf")
plt.title("Raw plotting of free energies of each window")
```

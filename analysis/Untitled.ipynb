```python
import numpy as np 
import matplotlib.pyplot as plt
mag=np.loadtxt("../output/tps_N.txt")
```

```python
mag1 = mag.reshape([-1, 150])
```

```python
plt.hist(mag1[20], bins=40)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


fig=plt.figure()
for m,num in zip(mag1, range(1,25)):
    ax = fig.add_subplot(5,5, num)
    ax.hist(m, bins=30)

plt.tight_layout()
plt.style.use('dark_background')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


fig=plt.figure(figsize=(10, 10))
for m,num in zip(mag1, range(1,25)):
    ax = fig.add_subplot(5,5, num)
    ax.hist(m, bins=30)

plt.tight_layout()
plt.savefig("fig.pdf")
```
